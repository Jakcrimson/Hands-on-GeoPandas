### Load our data
We'll be using one of the prebuilt datasets of geopandas : naturalearth_lowres (contour of countries)

```python
# Import the necessary libraries
import geopandas as gpd
import shapely.geometry
import matplotlib.pyplot as plt

# Initialize our geodataframe (gdf)
path_to_data = gpd.datasets.get_path("naturalearth_lowres")
gdf = gpd.read_file(path_to_data)

gdf.set_index('name')
```

Like the previous notebooks, for each country we have a a certain numer of attributes (estimated population, continent, gdp, name, geometry etc...)

### Visualize the data

We're going to plot the data based on the country's gdp_md_est attribute : 


```python
gdf["gdp_md_est"] = gdf.gdp_md_est
gdf.plot("gdp_md_est", legend=True, cmap="OrRd")
```

Okay, so the choice of color isn't the best one. However we can clearly see the countries with the highest estimated gdp (USA, CHINA) and what seems to be Alaska ? It's odd but we'll try to look for another choice of color palette :

```python
gdf["gdp_md_est"] = gdf.gdp_md_est
gdf.plot("gdp_md_est", legend=True, cmap="plasma", figsize=(20, 20))
```

The plasma color palette is a great choice when the granularity of the data is not very good. Now we have a better account of the countries and their respective gdp.

Now let's try to plot the countries regarding the continent they are on :

```python
gdf["continent"] = gdf.continent
gdf.plot("continent", legend=True, cmap="plasma", figsize=(20, 20))
```

Finally we'll plot the data regarding the number of people living in the country (pop_est) :

```python
gdf["pop_est"] = gdf.pop_est
gdf.plot("pop_est", legend=True, cmap="plasma", figsize=(10, 10))
```

So the result seems quite obivous, CHINA has the most import estimated population. Let's see the values for china :

```python
print("Max Value pop_est (CHINA) : ", gdf["pop_est"].max())
```
